# UNSUPERVISED BOUNDING BOXES FOR TRICOLOR IMAGES

## Importing the images
The inpute are binary images:

+ black = background
+ light blue = anomalies with 0.5 weight
+ dark blue = anomalies with 1.0 weight

The validation's bounding boxes are made at a guess. The pink box denotes the bounding box regarding to the whole anomaly, more or less.

```python
import matplotlib.pyplot as plt
import cv2
# import matplotlib.image as mpimg

# UPLOAD A RANDOM PICTURE
img_train_path = "train_tricolor_image.png"
img_val_path = "validation_tricolor_image.png"

# Display the images
img_train = cv2.imread(img_train_path)
img_val = cv2.imread(img_val_path)

rgb_train = cv2.cvtColor(img_train, cv2.COLOR_BGR2RGB)
rgb_val = cv2.cvtColor(img_val, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20, 18), dpi=80)
plt.imshow(rgb_train)
plt.title('train image')
plt.show()

plt.figure(figsize=(20, 18), dpi=80)
plt.imshow(rgb_val)
plt.title('validation image')
plt.show()
```

## Contouring

```python
## drawing all contours

image= img_train.copy()
gray= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

edges= cv2.Canny(gray,30,200)
contours, hierarchy= cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE) # RETR_TREE for the complete info in hierarchy

cv2.drawContours(image, contours, -1, (0,255,0),3)

rgb_train_box =  cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20, 18), dpi=80)
plt.imshow(rgb_train_box)
plt.title('Contours')
plt.show()
```

## Bounding box 

```python
img_bbox = img_train.copy()
boxes = []
for cntr in contours:
    x,y,w,h = cv2.boundingRect(cntr)
    boxes.append((x,y,w,h))
    cv2.rectangle(img_bbox, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
print('Box features:', boxes)
  

rgb_train_box =  cv2.cvtColor(img_bbox, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20, 18), dpi=80)
plt.imshow(rgb_train_box)
plt.title('bounding boxes')
plt.show()
```

## Hierarchy of the bounding boxes

```python
# Understanding hierarchy
print(hierarchy)

# Row 0 indicates [4 -1  1 -1]:
#     Next Contour: Contour 4
#     Previous Contour: -1 (Doesn’t exist)
#     First Child Contour: -1 (Doesn’t exist)
#     Parent Contour: -1 (Doesn’t exist)

# Row 1 indicates [-1 -1  2  0]:

#     Next Contour: Contour -1 (Doesn’t exist)
#     Previous Contour: Contour -1 (Doesn’t exist)
#     First Child Contour: 2 (Doesn’t exist)
#     Parent Contour: -1 (Doesn’t exist)

# Row 2 indicates [-1 -1  3  1]:

#     Next Contour: Contour -1 (Doesn’t exist)
#     Previous Contour: Contour -1 (Doesn’t exist)
#     First Child Contour: 3
#     Parent Contour: 1

```

```python
# We will use the hierarchy outputed by cv2.findContours to weight the inner geometries by 1 and their parents by 0.5
hierarchy_ = hierarchy[0] # get the actual inner list of hierarchy descriptions

child_box = [] # light blue
parent_box = [] # dark blue

# For each contour, find the bounding rectangle and draw it
for component in zip(contours, hierarchy_):
    currentContour = component[0]
    currentHierarchy = component[1]
    x,y,w,h = cv2.boundingRect(currentContour)
    if currentHierarchy[2] < 0:
        child_box.append((x,y,w,h))
        # these are the innermost child components
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    elif currentHierarchy[3] < 0:
        parent_box.append((x,y,w,h))
        # these are the outermost parent components
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)

print('Parent_boxes:', parent_box)
print('Child_boxes: ', child_box)

# Finally show the image
rgb_train_box_hierarchy =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20, 18), dpi=80)
plt.imshow(rgb_train_box_hierarchy)
plt.title('bounding boxes')
plt.show()
```

## Fusion of the bounding boxes depending on the weight

using: https://github.com/ZFTurbo/Weighted-Boxes-Fusion

```python
# Coordinates for boxes expected to be normalized e.g in range [0; 1]. Order: x1, y1, x2, y2.
def yolobbox2bbox(x,y,w,h):
    x1, y1 = x-w/2, y-h/2
    x2, y2 = x+w/2, y+h/2
    return x1, y1, x2, y2

h_img, w_img, _ = img.shape

b2b_parent_norm = []
b2b_child_norm = []
for box_p, box_c in zip(parent_box, child_box):
    # Parent
    # Normalize box coordinates
    x,y,w,h = box_p
    x /= w_img
    w /= w_img
    y /= h_img
    h /= h_img
    # Change format
    b2b_parent_norm.append(yolobbox2bbox(x,y,w,h))
    
    # Children
    # Normalize box coordinates
    x,y,w,h = box_c
    x /= w_img
    w /= w_img
    y /= h_img
    h /= h_img
    # Change format
    b2b_child_norm.append(yolobbox2bbox(x,y,w,h))
    

from ensemble_boxes import *

boxes_list = [b2b_parent_norm, b2b_child_norm]
print('boxes_list:', boxes_list)
weights = [2, 1]
scores_list = [[1.0, 1.0], [1.0, 1.0]] # The find contour are perfect
labels_list = [[0,1], [0, 1]] # Label of each object 

iou_thr = 0.5
skip_box_thr = 0.0001
sigma = 0.1


# Merge boxes for single model predictions
merged_boxes, merged_scores, merged_labels = weighted_boxes_fusion(boxes_list, scores_list, labels_list, weights=weights, iou_thr=iou_thr, skip_box_thr=skip_box_thr)
print(merged_boxes)
print(merged_scores)
print(merged_labels)
```

## Painting the merged box

```python
# Unnormalize the merged_boxes coordinates
def convert_yolobbox2bbox(size, box):
    dw = size[0]
    dh = size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh
    return (x,y,w,h)

# merged_boxes_x1x2 = []
# for box in merged_boxes:
#     # Change format
#     merged_boxes_x1x2.append(convert_yolobbox2bbox((w_img, h_img), box))
    
# print(merged_boxes_x1x2)

img_final = img_train.copy()
for box in merged_boxes:
    # Change format
    x, y, w, h = convert_yolobbox2bbox((w_img, h_img), box)
    print(x, y, w, h)
    # Draw bounding boxes
    cv2.rectangle(img_final,(x,y),(x+w,y+h),(0,255,0),2)


```
