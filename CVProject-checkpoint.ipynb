```python
# train folder
# train 1-0 means training data for 1 bounding box, 1st sample
# train 1-1 means training data for 1 bounding box, 2nd sample
# train 1-15 means training data for 1 bounding box, 16th sample
# train 2-0 means training data for 2 bounding boxes, 1st sample

# each sample has 70 images, B01 is center camera, B01,B02,..,B05,G01,G02,..G05 = 10 cameras, 7 frames

# each sample has homographies.json: Homographie 3x3 matrix for all RGB images w.r.t center camera 3_B01

# validation folder: same as train folder but in addition labels.json for the bounding boxes

# bounding boxes are from 3_B01 camera
```

```python
import utils
import cv2
import imutils
import glob
import os
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
# Read train images
datafolder = 'data_WiSAR/data/train/'

dirs = os.listdir(datafolder)

raw_train_img = dict()
homography_files = []
for act_dir in dirs:
    raw_train_img[act_dir] = dict()
    samples = glob.glob(datafolder+act_dir+'/*.png')
    for img in samples:
        file = os.path.basename(img)
        img_name = os.path.splitext(file)[0]
        #print('Processing ',img)
        src = cv2.cvtColor(cv2.imread(img),cv2.COLOR_BGR2RGB) ##Read images using imread function and convert color space using cvtColor
        raw_train_img[act_dir][img_name]=src
    # read in homographies.json
    f = glob.glob(datafolder+act_dir+'/*.json')
    homography_files.append(f)

print(len(raw_train_img))
```

```python
raw_train_img['train-1-0']['0-B01'].shape
```

```python
with open(str(np.squeeze(homography_files[0])), "r") as read_file:
    homographies = json.load(read_file)
homography = np.array(homographies['0-B01'])
image = raw_train_img['train-1-0']['0-B01']
print(image.shape)
warped_image = cv2.warpPerspective(image,homography,image.shape[:2])
```

```python
print(warped_image.shape)
plt.figure()
plt.imshow(warped_image)
plt.show()
```
